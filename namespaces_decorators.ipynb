```python
# scope is a textual region of a python program where namespace is directly accessible.
#LEGB Rule
```

```python
#local and global

#global scope
a = 2

def temp():
  #local scope
  b=3
  print(b)

temp()
print(a)
```

```python
#global scope
a = 2

def temp():
  #local scope
  a=3
  print(a)

temp()
print(a)
```

```python
#local and global -> local does not have but global has

#global
a = 2

def temp():
  #local scope
  print(a)

temp()
print(a)
```

```python
#local and global -> editing global

#global

try:
  a = 2
  def temp():
    #local scope
    a+=1
    print(a)

  temp()
  print(a)
except:
  print('We can access global scope variable in local but can not change it in local scope.')


```

```python
#Using global keyword
try:
  a = 2
  def temp():
    #local scope
    global a
    a+=1
    print(a)

  temp()
  print(a)
  print('We can edit global variable in local by using global keyword')
except:
  print('We can access global scope variable in local but can not change it in local scope.')
```

```python
#local and global -> global created inside local

def temp():
  #local scope
  global a
  a=3
  print(a)

temp()
print(a)
print('we can create variable in local scope for global purpose also')
```

```python
#local and global -> function parameter is local

def temp(z):
  #local scope
  print(z)

a=5
temp(a)
print(a)
```

```python
#enclosing scope: seen in nested function

def outer():
  def inner():
    print('inner')
  print('outer')
  inner()
outer()
print('Main')
```

```python
def outer():
  a = 3
  def inner():
    print(a)
  print(a)
  inner()
outer()
print('Main')
```

```python
def outer():
  a = 3
  def inner():
    nonlocal a
    a+=1
    print(a)
  print(a)
  inner()
outer()
print('Main')
```

```python
#Decorators : A decorators in python is a function that receives another function as input and adds some functionality(decoration) to it and returns it.
#This can happen only because python functions are first class citizen.

#Two types of decorators:
#Built in decorators : @staticmethod, @classmethod, @abstractmethod, and @propertymethod
#User defined decorators: programmers can create according to it.
```

```python
#functions are first class citizen
def func():
  print('Hello')

a = func
a()
```

```python
def modify2(func,num):
  return func(num)

def square(num):
  return num**2

modify2(square,2)
```

```python
#simple decorators
def my_decorator(func):
  def wrapper():
    print('*******')
    func()
    print('*******')
  return wrapper

def hello():
  print('Hello')

a = my_decorator(hello)
a()
```

```python
def display():
  print('Hello Chayan')

b = my_decorator(display)
b()
```

```python
def outer():
  a = 5
  def inner():
    print(a)
  return inner

a = outer()
a()
```

```python
#simple decorators
def my_decorator(func):
  def wrapper():
    print('*******')
    func()
    print('*******')
  return wrapper

@my_decorator
def hello():
  print('Hello')

hello()
```

```python
import time

def timer(func):
  def wrapper():
    start = time.time()
    func()
    print('time taken by :',func.__name__,time.time()-start,' sec')

  return wrapper

@timer
def hello():
  print('Hello')
  time.sleep(2)

@timer
def display():
  print('Display')
  time.sleep(4)

hello()
display()
```

```python
import time

def timer(func):
  def wrapper(*args):
    start = time.time()
    func(*args)
    print('time taken by :',func.__name__,time.time()-start,' sec')

  return wrapper

@timer
def square(num):
  time.sleep(2)
  print(num**2)

square(2)
```

```python
@timer
def power(a,b):
  print(a**b)

power(3,2)
```

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(args[0]) == data_type:
        func(*args)
      else:
        raise TypeError('Invalid Type')
    return inner_wrapper # Return the inner wrapper function
  return outer_wrapper

@sanity_check(int)
def square(i):
  print(i**2)

square(2)
```
