# Text Preprocessing

```python
import string
# removing the html tags
def clean_html(text):
    clean=re.compile('<.*?>')
    cleantext=re.sub(clean,'',text)
    return cleantext

# first round of cleaning
def clean_text1(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text=re.sub('\w*\d\w*','',text)
    return text

# second round of cleaning
def clean_text2(text):
    text=re.sub('[''"",,,]','',text)
    text=re.sub('\n','',text)
    return text

cleaned_html=lambda x:clean_html(x)
cleaned1=lambda x:clean_text1(x)
cleaned2=lambda x:clean_text2(x)

df1['review']=pd.DataFrame(df1.review.apply(cleaned_html))
df1['review']=pd.DataFrame(df1.review.apply(cleaned1))
df1['review']=pd.DataFrame(df1.review.apply(cleaned2))
df1.head()
```

```python
df1.value_counts('sentiment')
```

```python
X = df1.iloc[:,0:1]
y = df1['sentiment']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Applying BOW
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_train_bow.shape
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
gnb = GaussianNB()
gnb.fit(X_train_bow,y_train)
```

```python
#Gaussian Naive Bayes
y_pred = gnb.predict(X_test_bow)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#Multinomial Naive Bayes (used for text data)
mnb = MultinomialNB()
mnb.fit(X_train_bow,y_train)

y_pred1 = mnb.predict(X_test_bow)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
#applying Random forest
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_bow,y_train)

y_pred2 = rfc.predict(X_test_bow)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
#Using most common words in BOW
cv = CountVectorizer(max_features=3000)
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()
rf.fit(X_train_bow,y_train)

y_pred3 = rf.predict(X_test_bow)
print(accuracy_score(y_test,y_pred3))
```

```python
#Using n_gram

cv = CountVectorizer(ngram_range=(1,3),max_features=5000)

X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()
rf.fit(X_train_bow,y_train)

y_pred4 = rf.predict(X_test_bow)
print(accuracy_score(y_test,y_pred4))
```

```python
#Applying TF-IDF (Used for information retrival system)

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train['review']).toarray()
X_test_tfidf = tfidf.transform(X_test['review']).toarray()

rf = RandomForestClassifier()
rf.fit(X_train_tfidf,y_train)

y_pred5 = rf.predict(X_test_tfidf)
print(accuracy_score(y_test,y_pred5))
```

```python
#Applying word2vec

import gensim
from gensim.models import Word2Vec
from nltk import sent_tokenize,word_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')

story = []
for doc in df1['review']:
    raw_sent = sent_tokenize(doc)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))

model = Word2Vec(story,min_count=2,window = 10)

model.build_vocab(story)
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
len(model.wv.index_to_key)
```

```python
import numpy as np
def document_vector(doc):
  #removing out of vocab words
  doc = [word for word in doc.split() if word in model.wv.index_to_key]
  return np.mean(model.wv[doc],axis=0)
```

```python
document_vector(df1['review'].values[0])
```

```python
from tqdm import tqdm
X = []
for doc in tqdm(df1['review'].values):
  X.append(document_vector(doc))
```
