# Pricing and Billing using Boto3

```python
#Cost Explorar

client = boto3.client('ce')
#Start time
from datetime import datetime, timedelta
start_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d') #CE-- start-date --> end date

end_date = datetime.now().strftime('%Y-%m-%d')

response = client.get_cost_and_usage(TimePeriod={'Start':start_date,
                                                'End':end_date},
                                    Granularity='MONTHLY',
                                    Metrics=['UnblendedCost','UsageQuantity']) #granularity : DAILY, MONTHLY, HOURLY

for item in response['ResultsByTime']:
  # Extract start date and cost
  start_date = item['TimePeriod']['Start']
  cost = item['Total']['UnblendedCost']['Amount']
  print(start_date, cost)  # Print start date and cost
```

```python
for item in response['ResultsByTime']:
  print(item['TimePeriod'])
  print(item['Total']['UnblendedCost'])
  print('\n\n')
```

```python
#what services we have used
response = client.get_dimension_values(TimePeriod={'Start':start_date,
                                                'End':end_date},
                                    Dimension='SERVICE')
for service in response['DimensionValues']:
  print(service['Value'])
```

```python
#how much money spend per service (Using groupby)

response = client.get_cost_and_usage(
    TimePeriod={'Start':start_date,
                'End':end_date},
    Granularity='MONTHLY',
    Metrics=['UnblendedCost','UsageQuantity'],
    GroupBy = [{
        'Type':'DIMENSION',
        'Key':'SERVICE'
    }]
)

for item in response['ResultsByTime']:
  print(item['TimePeriod'])
  print('\n')
  for group in item['Groups']:
    service_name = group['Keys'][0]
    cost = group['Metrics']['UnblendedCost']['Amount']
    print(f'{service_name}: ${cost}')
  print('\n\n')
  # print(item['Groups'][0]['Keys'][0], item['Groups'][0]['Metrics']['UnblendedCost']['Amount'])
```

```python
#Future Prediction of how mutch time spend
start_time = (datetime.now() + timedelta(days=1)).strftime('%Y-%m-%d')
end_time = (datetime.now()+timedelta(days=31)).strftime('%Y-%m-%d')

result = client.get_cost_forecast(TimePeriod={'Start':start_time,
                                    'End':end_time},
                        Granularity='MONTHLY',
                        Metric='UNBLENDED_COST')

result

```

```python
import boto3 # Import the boto3 library

# Create a DynamoDB client
client = boto3.client('dynamodb', region_name='us-east-1')

entry = {
    'Title': {'S': 'The Shawshank Redemption'}, # Use a colon to separate the attribute type and value
    'Director': {'S': 'Frank Darabont'},
    'Year': {'N': '1994'},
    'Rating': {'N': '9.3'}
}
# Call the put_item method on the DynamoDB client
client.put_item(TableName='Movies',Item=entry)

item_key = {
    'Title':{'S': 'The Shawshank Redemption'},
    'Rating':{'N': '9.3'}
}
response = client.get_item(TableName='Movies',Key=item_key)
response
```

```python
response['Item']
```

```python
#Updating item in the table
update = 'SET Director = :r'
client.update_item(TableName='Movies',
                   Key=item_key,
                   UpdateExpression=update,
                   ExpressionAttributeValues={':r': {'S': 'Francis Ford Coppola'}})
```
