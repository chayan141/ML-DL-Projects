```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('/content/macrodata.csv',index_col=0,parse_dates=True)
```

```python
df.head()
```

```python
df['realgdp'].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter #hodric prescot filter
```

```python
gdp_cycle , gdp_trend = hpfilter(df['realgdp'],lamb=1600)
```

```python
type(gdp_trend)
```

```python
gdp_trend.plot()
```

```python
df['trend'] = gdp_trend
```

```python
df.head()
```

```python
df[['trend','realgdp']]['2005-01-01':].plot(figsize=(12,5))
```

```python
airline = pd.read_csv('/content/airline_passengers.csv',index_col = 'Month',parse_dates=True)
```

```python
airline
```

```python
airline = airline.dropna()
```

```python
airline.head()
```

```python
airline.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model = 'multiplicative')
```
