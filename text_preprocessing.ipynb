```python
#removing urls
import re
import pandas as pd

# Function to remove URLs from text
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Sample DataFrame
data = {
    'text': [
        'Visit our site at https://www.example.com!',
        'Check out this post http://example.org/post/1234',
        'No URLs here, just plain text.',
        'Another link: www.example.com/some-page'
    ]
}
df = pd.DataFrame(data)

# Apply the function to remove URLs
df['clean_text'] = df['text'].apply(remove_urls)

# Display the DataFrame
df.head()
```

```python
#using nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Example usage
text = "This is a simple example to demonstrate the removal of stopwords."
cleaned_text = remove_stopwords(text)
print(cleaned_text)  # Output: "simple example demonstrate removal stopwords"
```

```python
#using spacy
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

def remove_stopwords(text):
    doc = nlp(text)
    filtered_words = [token.text for token in doc if not token.is_stop]
    return ' '.join(filtered_words)

# Example usage
text = "This is a simple example to demonstrate the removal of stopwords."
cleaned_text = remove_stopwords(text)
print(cleaned_text)  # Output: "simple example demonstrate removal stopwords"
```

```python
#removing emojis
import re

def remove_emojis(text):
    # Define the regex pattern for emojis
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U00002702-\U000027B0"  # other symbols
        "\U000024C2-\U0001F251"
        "\U00002500-\U00002BEF"  # Chinese characters
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Example usage
text_with_emojis = "I love Python ! It's so cool . Let's code !"
cleaned_text = remove_emojis(text_with_emojis)
print(cleaned_text)  # Output: "I love Python ! It's so cool . Let's code !"
```

Stemming

In grammer, Inflection is the modification of a word to express different grammatical categories such as tense, case, voice, aspect, person, number, gender & mood.

Stemming brings inflected word in its root form. Used is informtation retrival system. It involves removing prefixes and suffixes from a word. In nlp, stemming helps to normalise words that have similar meaning but different in form.

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Example text
text = "running runs ran runner"

# Tokenize the text into words
words = word_tokenize(text)

# Perform stemming
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

```python
sample = ['walked', 'walks', 'walking']
for word in sample:
  print(stemmer.stem(word))
```

```python
#using spacy for stemming
import spacy

# Load the English language model
nlp = spacy.load('en_core_web_sm')

# Example text
text = "running runs ran runner"

# Process the text with spaCy
doc = nlp(text)

# Lemmatization with spaCy
lemmatized_words = [token.lemma_ for token in doc]

print(lemmatized_words)
```

Tha goal of lemmagtization is to normalise words while still ensuring that the reduced form belongs to the language and maintained its meaning. Unlike stemming, which involves chopping off prefixes and suffixes without considering the context of the word, lemmtisation takes into account the context and parts of speech of the word in the sentence.

```python
#lammatization using NLTK(Natural Language Toolkit)
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize words
words = ["running", "better", "mice"]
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print(lemmatized_words)
```
