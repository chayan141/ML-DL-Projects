```python
#list travarsing : process of accessing each elements in a list. Iterating through the list
```

```python
#list traversing using for loop
my_list = [1, 2, 3, 4, 5]
for element in my_list:
    print(element)
```

```python
my_list = [1, 2, 3, 4, 5]
for i in range(len(my_list)):
    print(my_list[i])
```

```python
#using while loop
my_list = [1, 2, 3, 4, 5]
i = 0
while i < len(my_list):
    print(my_list[i])
    i += 1
```

```python
#using list comprehention
my_list = [1, 2, 3, 4, 5]
squared_list = [x ** 2 for x in my_list]
print(squared_list)
```

```python
#using enumerate : enumerate() is useful when you need both the index and the value during traversal.
for index, value in enumerate(my_list):
    print(f"Index: {index}, Value: {value}")
```

```python
#Using Map Function : It applies a specific function to each items in an iterable like list and returns a map object
squared_list = list(map(lambda x: x ** 2, my_list))
print(squared_list)
```

```python
#Filter function: It filters the list by appling a function in each elements in that list and returns true or false for each element.
my_list = [1, 2, 3, 4, 5]
even_list = list(filter(lambda x: x % 2 == 0, my_list))
print(even_list)
```

```python
#reduce method: From the functools module, reduce() applies a rolling computation to the items of the list.

#adding all elemets of a list
from functools import reduce
my_list = [1, 2, 3, 4, 5]
sum_list = reduce(lambda x, y: x + y, my_list)
print(sum_list)
```

```python
#multipling all elements in a list
my_list = [1, 2, 3, 4, 5]
mul = reduce(lambda x,y:x*y,my_list)
print(mul)
```

```python
#zip() : It adds multiple iterables together and convert them into a single tuple.
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

paired_list = zip(names, ages)
print(list(paired_list))
```

```python
#Unzipping
combined = [('Alice', 25), ('Bob', 30), ('Charlie', 35)]

names, ages = zip(*combined)
print(names)  # ('Alice', 'Bob', 'Charlie')
print(ages)   # (25, 30, 35)
```

```python
#iterating multiple list together
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

for num, letter in zip(list1, list2):
    print(f"Number: {num}, Letter: {letter}")
```

```python
#converting into dictionary
keys = ['name', 'age', 'city']
values = ['Alice', 25, 'New York']

dictionary = dict(zip(keys, values))
print(dictionary)
```

```python
#multiple list comprehention
list1 = [1, 2, 3]
list2 = [4, 5, 6]

summed_list = [x + y for x, y in zip(list1, list2)]
print(summed_list)
```

```python
my_dict = {(1, 2): "a pair", (3, 4): "another pair"}
print(my_dict[(1, 2)])
```

```python
tuple1 = (1, 2, 3)
tuple2 = ('a', 'b', 'c')

zipped = zip(tuple1, tuple2)
print(list(zipped))
```

```python
zipped_tuple = ((1, 'a'), (2, 'b'), (3, 'c'))

# Unzipping using the zip(*) trick
unzipped = zip(*zipped_tuple)
tuple1, tuple2 = map(tuple, unzipped)  # Convert the unzipped parts back to tuples

print(tuple1)  # Output: (1, 2, 3)
print(tuple2)  # Output: ('a', 'b', 'c')
```

```python
#frozen set : immutable set
my_frozenset = frozenset([1, 2, 3, 4])
print(my_frozenset)
```

```python
d = {
    frozenset([1, 2, 3]): "group1",
    frozenset([4, 5, 6]): "group2"
}

print(d)  # Output: {frozenset({1, 2, 3}): 'group1', frozenset({4, 5, 6}): 'group2'}
print(d[frozenset([1, 2, 3])])  # Output: group1
```

```python
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])

# Union
print(fs1.union(fs2))
```

```python
# Intersection
print(fs1.intersection(fs2))
```

```python
# Difference
print(fs1.difference(fs2))
```

```python
# Symmetric Difference
print(fs1.symmetric_difference(fs2))
```

```python
# Subset and Superset
print(fs1.issubset(fs2))  # Output: False
print(fs1.issuperset(fs2))
```

```python
# Disjoint Sets
print(fs1.isdisjoint(fs2))
```

Aspects of Univariate Analysis :
Descriptive Statistics: To summarize and describe the main features of a dataset.
Distribution Understanding: To understand the spread and shape of the data.
Outlier Detection: To identify any unusual or extreme values in the data.
Data Quality Assessment: To assess the quality of the data, including missing values and potential errors.

Central Tendency: Describes the center of the data distribution. Mean, Median, Mode

Dispersion : How spread out the values are around the central tendency.

Range: Difference between maximum and minimum values.

Variance : Average of squared difference from mean.

SD : The avverage amount of values deviated frommean.

Interquartile Range : Range Between 1st quartile (25th Percentile) and third quartile (75th Percentile).

Distribution Shape: Histogram, Boxplot, Density Plot, Skewness, Kurtosis

Skewness : A measure of asymmetry of the data distribution

Kurtosis: A measure of the tailedness  of  the data distribution.

Outliers : Boxplot Analysis : points outside of the whiskers, typically 1.5 times the IQR from the quartiles.

Z  score : Datapoints wit a z score(numbber of SD from the mean) greater than the threshold(3) are considered as outliers.




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())

# Choose a variable for univariate analysis
variable = 'sepal_length'

# Descriptive statistics
mean_value = iris[variable].mean()
median_value = iris[variable].median()
mode_value = iris[variable].mode()[0]
std_deviation = iris[variable].std()
range_value = iris[variable].max() - iris[variable].min()
iqr_value = iris[variable].quantile(0.75) - iris[variable].quantile(0.25)
skewness_value = iris[variable].skew()
kurtosis_value = iris[variable].kurtosis()

# Print statistics
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
print(f"Standard Deviation: {std_deviation}")
print(f"Range: {range_value}")
print(f"IQR: {iqr_value}")
print(f"Skewness: {skewness_value}")
print(f"Kurtosis: {kurtosis_value}")
```

```python
# Set the size of the plots
plt.figure(figsize=(14, 6))

# Histogram with KDE (Kernel Density Estimate)
plt.subplot(1, 2, 1)
sns.histplot(iris[variable], bins=30, kde=True, color='blue')
plt.title(f'Histogram of {variable.capitalize()}')
plt.xlabel(variable.capitalize())
plt.ylabel('Frequency')

# Box plot
plt.subplot(1, 2, 2)
sns.boxplot(x=iris[variable], color='green')
plt.title(f'Box Plot of {variable.capitalize()}')
plt.xlabel(variable.capitalize())

# Display the plots
plt.tight_layout()
plt.show()
```
