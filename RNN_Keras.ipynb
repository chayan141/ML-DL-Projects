```python
from keras.datasets import imdb
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
import numpy as np
np.array(X_train).shape
```

```python
X_test.shape
```

```python
sent = [
    'India won the world',
    'Rohit played very well',
    'Above all is my imagination',
    'India is great',
    'definately we win next worldcup'
]
```

```python
from keras.preprocessing.text import Tokenizer #performs integer incoding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sent)
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
encoding = tokenizer.texts_to_sequences(sent)
```

```python
#converting all sentences into same length
from keras.utils import pad_sequences
pad_sequences(encoding,padding='post')
```

```python
X_train
```

```python
mylist=[]
for i in X_train:
  mylist.append(len(i))
min(mylist)
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
model = Sequential()
model.add(SimpleRNN(32,input_shape=(50,1)))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5)
```

```python
model.predict(X_test)
```

```python
model.evaluate(X_test,y_test)
```

```python
loss,accuracy = model.evaluate(X_test,y_test)
loss
```
